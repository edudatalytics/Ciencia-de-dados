```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib notebook 
%matplotlib inline
```

```python
sinasc = pd.read_csv(r"C:\Users\eduar\Desktop\SINASC_RO_2019.csv")
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
# Criar subplots para cada variável
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Gráficos de Barras para Variáveis Selecionadas', fontsize=16)

# Gráfico de Barras para APGAR1
sinasc['APGAR1'].value_counts().sort_index().plot(kind='bar', ax=axs[0, 0], color='skyblue', edgecolor='black')
axs[0, 0].set_title('APGAR1')
axs[0, 0].set_xlabel('Pontuação APGAR1')
axs[0, 0].set_ylabel('Número de Ocorrências')

# Gráfico de Barras para APGAR5
sinasc['APGAR5'].value_counts().sort_index().plot(kind='bar', ax=axs[0, 1], color='lightgreen', edgecolor='black')
axs[0, 1].set_title('APGAR5')
axs[0, 1].set_xlabel('Pontuação APGAR5')
axs[0, 1].set_ylabel('Número de Ocorrências')

# Gráfico de Barras para SEXO
sinasc['SEXO'].value_counts().plot(kind='bar', ax=axs[0, 2], color='lightcoral', edgecolor='black')
axs[0, 2].set_title('SEXO')
axs[0, 2].set_xlabel('Sexo do Bebê')
axs[0, 2].set_ylabel('Número de Ocorrências')

# Gráfico de Barras para GESTACAO
sinasc['GESTACAO'].value_counts().sort_index().plot(kind='bar', ax=axs[1, 0], color='gold', edgecolor='black')
axs[1, 0].set_title('GESTACAO')
axs[1, 0].set_xlabel('Idade Gestacional')
axs[1, 0].set_ylabel('Número de Ocorrências')

# Gráfico de Barras para ESTCIVMAE
sinasc['ESTCIVMAE'].value_counts().plot(kind='bar', ax=axs[1, 1], color='lightblue', edgecolor='black')
axs[1, 1].set_title('ESTCIVMAE')
axs[1, 1].set_xlabel('Estado Civil da Mãe')
axs[1, 1].set_ylabel('Número de Ocorrências')

# Remover o subplot não utilizado
fig.delaxes(axs[1, 2])

# Ajustar o layout
plt.tight_layout(rect=[0, 0, 1, 0.96])

# Exibir os gráficos
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
# Calcular a média do APGAR5 por número de consultas
media_apgar5_por_consultas = sinasc.groupby('CONSULTAS')['APGAR5'].mean()

# Criar um gráfico de barras
plt.figure(figsize=(10, 6))
media_apgar5_por_consultas.plot(kind='bar', color='lightcoral', edgecolor='black')

# Adicionar rótulos ao gráfico
plt.title('Média do APGAR5 por Número de Consultas Pré-Natais')
plt.xlabel('Número de Consultas Pré-Natais')
plt.ylabel('Média do APGAR5')

# Exibir o gráfico
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
# Criar um painel com 2 linhas e 2 colunas de subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Distribuição de Frequência por Variável', fontsize=16)

# Gráfico de barras para SEXO
sns.countplot(x='SEXO', data=sinasc, ax=axs[0, 0], palette='viridis')
axs[0, 0].set_title('Distribuição de Frequência de SEXO')

# Gráfico de barras para GESTACAO
sns.countplot(x='GESTACAO', data=sinasc, ax=axs[0, 1], palette='plasma')
axs[0, 1].set_title('Distribuição de Frequência de GESTACAO')

# Gráfico de barras para ESTCIVMAE
sns.countplot(x='ESTCIVMAE', data=sinasc, ax=axs[1, 0], palette='mako')
axs[1, 0].set_title('Distribuição de Frequência de ESTCIVMAE')

# Gráfico de barras para CONSULTAS
sns.countplot(x='CONSULTAS', data=sinasc, ax=axs[1, 1], palette='rocket')
axs[1, 1].set_title('Distribuição de Frequência de CONSULTAS')

# Ajustar o layout
plt.tight_layout(rect=[0, 0, 1, 0.96])

# Exibir os gráficos
plt.show()
```
